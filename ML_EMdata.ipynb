```python
# Synthetic TDEM dataset generator and visualization
# This cell will:
# 1) Generate 1,000 synthetic TDEM decay curves (48 time gates, log-spaced from 10 µs to 10 ms)
# 2) Mark 5% (50 curves) as "noisy" with heavy disturbances
# 3) Save the dataset to CSV and Parquet
# 4) Plot samples of normal vs noisy curves
# 5) Create a self-contained Jupyter notebook file (.ipynb) with the same code
#
# Notes:
# - Plots follow the requirements: matplotlib only, one chart per figure, no explicit colors.
# - Y-axis is linear (values may include noise; log-y would not allow non-positive values).
# - X-axis uses log-scale to reflect gate timing.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import os

# Reproducibility
rng = np.random.default_rng(42)

# -----------------------------
# 1) Define time gates
# -----------------------------
N_CURVES = 1000
NOISY_FRAC = 0.05
N_NOISY = int(N_CURVES * NOISY_FRAC)
N_GATES = 48

# 10 microseconds to 10 milliseconds
t_sec = np.logspace(-5, -2, N_GATES)  # seconds
t_ms = t_sec * 1e3                    # milliseconds (for plotting and saving)
gate_cols = [f"gate_{i+1}" for i in range(N_GATES)]

# -----------------------------
# 2) Synthetic "clean" decay model
# S(t) = A * (t + t0)^(-beta) * exp(-t / tau) + small noise/spikes
# -----------------------------
def synth_clean_decay(t, rng):
    # Amplitude in arbitrary units
    A = 10 ** rng.uniform(0.0, 1.0)         # ~ [1, 10]
    beta = rng.uniform(1.0, 2.5)            # power-law slope
    tau = 10 ** rng.uniform(-3.5, -2.3)     # ~ [0.000316, 0.005] s
    t0  = 10 ** rng.uniform(-6.0, -4.3)     # ~ [1e-6, 5e-5] s

    base = A * (t + t0) ** (-beta) * np.exp(-t / tau)

    # Small relative Gaussian noise proportional to early amplitude
    rel_std = rng.uniform(0.01, 0.03)       # 1–3% noise
    sigma = rel_std * np.maximum(np.median(base), 1e-8)
    noise = rng.normal(0.0, sigma, size=base.shape)

    # Rare small spikes (very low probability)
    spike_mask = rng.random(size=base.shape) > 0.997  # ~0.3% of gates
    spikes = spike_mask * rng.normal(loc=3*sigma, scale=1.0*sigma, size=base.shape)

    y = base + noise + spikes
    # Ensure non-negative small floor (not strictly required but helps visuals)
    y = np.maximum(y, 0.0)
    return y

# -----------------------------
# 3) Heavy noise injector (for 5%)
# -----------------------------
def inject_heavy_noise(y, t, rng):
    y = y.copy()

    # Add DC offset / baseline drift (slow sine)
    amp_drift = rng.uniform(0.05, 0.3) * max(np.median(y), 1e-8)
    drift = amp_drift * np.sin(2 * np.pi * rng.uniform(0.2, 1.0) * (t / t.max()))
    y += drift

    # Strong Gaussian noise
    strong_sigma = rng.uniform(0.2, 0.6) * max(np.median(y), 1e-8)
    y += rng.normal(0.0, strong_sigma, size=y.shape)

    # Add many spikes
    spike_prob = rng.uniform(0.05, 0.15)             # 5–15% gates spike
    spike_mask = rng.random(size=y.shape) < spike_prob
    spike_amp = rng.uniform(2.0, 6.0) * strong_sigma
    y += spike_mask * rng.normal(spike_amp, 0.5 * spike_amp, size=y.shape)

    # Random partial gate corruption (replace some gates with random noise)
    corrupt_prob = rng.uniform(0.05, 0.2)            # 5–20% gates corrupted
    corrupt_mask = rng.random(size=y.shape) < corrupt_prob
    y[corrupt_mask] = rng.normal(loc=0.0, scale=2.0 * strong_sigma, size=np.sum(corrupt_mask))

    # Clamp negatives to zero for stability of linear-scale plotting
    y = np.maximum(y, 0.0)
    return y

# -----------------------------
# 4) Generate dataset
# -----------------------------
curves = []
labels = np.zeros(N_CURVES, dtype=int)  # 0=normal, 1=noisy

for idx in range(N_CURVES):
    y = synth_clean_decay(t_sec, rng)
    curves.append(y)

# Pick noisy indices
noisy_indices = rng.choice(N_CURVES, size=N_NOISY, replace=False)
labels[noisy_indices] = 1

# Inject heavy noise into selected curves
for idx in noisy_indices:
    curves[idx] = inject_heavy_noise(curves[idx], t_sec, rng)

curves = np.array(curves)  # shape: (N_CURVES, N_GATES)

# Construct DataFrame
df = pd.DataFrame(curves, columns=gate_cols)
df.insert(0, "curve_id", np.arange(N_CURVES))
df["is_noisy"] = labels

# -----------------------------
# 5) Save dataset
# -----------------------------
os.makedirs("/mnt/data", exist_ok=True)
csv_path = "/mnt/data/tdem_synth.csv"
parquet_path = "/mnt/data/tdem_synth.parquet"
times_csv_path = "/mnt/data/tdem_times_ms.csv"

df.to_csv(csv_path, index=False)
try:
    df.to_parquet(parquet_path, index=False)
except Exception as e:
    # Parquet might not be available depending on environment; ignore if fails
    parquet_path = None

# Save times (ms) to CSV (first row contains gate times)
pd.DataFrame([t_ms], columns=gate_cols).to_csv(times_csv_path, index=False)

# -----------------------------
# 6) Visualizations
# -----------------------------

# Plot a sample of normal curves (overlay)
normal_idx = np.where(labels == 0)[0]
sample_normal = rng.choice(normal_idx, size=min(100, len(normal_idx)), replace=False)

plt.figure(figsize=(8, 5))
for idx in sample_normal:
    plt.plot(t_ms, curves[idx], alpha=0.4)
plt.xscale("log")
plt.xlabel("Time (ms)")
plt.ylabel("Amplitude (arb. units)")
plt.title("Sample of NORMAL synthetic TDEM decays (overlay)")
plt.tight_layout()
plt.show()

# Plot ALL noisy curves (5% -> about 50 curves)
plt.figure(figsize=(8, 5))
for idx in noisy_indices:
    plt.plot(t_ms, curves[idx], alpha=0.7)
plt.xscale("log")
plt.xlabel("Time (ms)")
plt.ylabel("Amplitude (arb. units)")
plt.title("NOISY synthetic TDEM decays (overlay)")
plt.tight_layout()
plt.show()

# Plot one normal vs one noisy for direct comparison
plt.figure(figsize=(8, 5))
ex_normal = rng.choice(normal_idx)
ex_noisy = rng.choice(noisy_indices)
plt.plot(t_ms, curves[ex_normal], label="normal", alpha=0.9)
plt.plot(t_ms, curves[ex_noisy], label="noisy", alpha=0.9)
plt.xscale("log")
plt.xlabel("Time (ms)")
plt.ylabel("Amplitude (arb. units)")
plt.title("Example: one normal vs one noisy curve")
plt.legend()
plt.tight_layout()
plt.show()

# -----------------------------
# 7) Show a preview of the dataset
# -----------------------------
print("Synthetic TDEM dataset (first 20 rows)")
display(df.head(20))
# -----------------------------
# 8) Create a self-contained .ipynb with the same generator code
# -----------------------------
nb_code = r'''# Synthetic TDEM dataset generator and visualization
# This notebook generates 1,000 synthetic TDEM decay curves, where 95% are normal and 5% contain heavy noise.
# It also saves the dataset to CSV/Parquet and plots example overlays.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

rng = np.random.default_rng(42)

# --- Time gates ---
N_CURVES = 1000
NOISY_FRAC = 0.05
N_NOISY = int(N_CURVES * NOISY_FRAC)
N_GATES = 48
t_sec = np.logspace(-5, -2, N_GATES)  # 10µs .. 10ms
t_ms = t_sec * 1e3
gate_cols = [f"gate_{i+1}" for i in range(N_GATES)]

def synth_clean_decay(t, rng):
    A = 10 ** rng.uniform(0.0, 1.0)         # [1, 10]
    beta = rng.uniform(1.0, 2.5)
    tau = 10 ** rng.uniform(-3.5, -2.3)
    t0  = 10 ** rng.uniform(-6.0, -4.3)
    base = A * (t + t0) ** (-beta) * np.exp(-t / tau)
    rel_std = rng.uniform(0.01, 0.03)
    sigma = rel_std * np.maximum(np.median(base), 1e-8)
    noise = rng.normal(0.0, sigma, size=base.shape)
    spike_mask = rng.random(size=base.shape) > 0.997
    spikes = spike_mask * rng.normal(loc=3*sigma, scale=1.0*sigma, size=base.shape)
    y = base + noise + spikes
    y = np.maximum(y, 0.0)
    return y

def inject_heavy_noise(y, t, rng):
    y = y.copy()
    amp_drift = rng.uniform(0.05, 0.3) * max(np.median(y), 1e-8)
    drift = amp_drift * np.sin(2 * np.pi * rng.uniform(0.2, 1.0) * (t / t.max()))
    y += drift
    strong_sigma = rng.uniform(0.2, 0.6) * max(np.median(y), 1e-8)
    y += rng.normal(0.0, strong_sigma, size=y.shape)
    spike_prob = rng.uniform(0.05, 0.15)
    spike_mask = rng.random(size=y.shape) < spike_prob
    spike_amp = rng.uniform(2.0, 6.0) * strong_sigma
    y += spike_mask * rng.normal(spike_amp, 0.5 * spike_amp, size=y.shape)
    corrupt_prob = rng.uniform(0.05, 0.2)
    corrupt_mask = rng.random(size=y.shape) < corrupt_prob
    y[corrupt_mask] = rng.normal(loc=0.0, scale=2.0 * strong_sigma, size=np.sum(corrupt_mask))
    y = np.maximum(y, 0.0)
    return y

curves = []
labels = np.zeros(N_CURVES, dtype=int)
for idx in range(N_CURVES):
    curves.append(synth_clean_decay(t_sec, rng))
noisy_indices = rng.choice(N_CURVES, size=N_NOISY, replace=False)
labels[noisy_indices] = 1
for idx in noisy_indices:
    curves[idx] = inject_heavy_noise(curves[idx], t_sec, rng)
curves = np.array(curves)

df = pd.DataFrame(curves, columns=gate_cols)
df.insert(0, "curve_id", np.arange(N_CURVES))
df["is_noisy"] = labels

os.makedirs("/mnt/data", exist_ok=True)
df.to_csv("/mnt/data/tdem_synth.csv", index=False)
try:
    df.to_parquet("/mnt/data/tdem_synth.parquet", index=False)
except Exception:
    pass
pd.DataFrame([t_ms], columns=gate_cols).to_csv("/mnt/data/tdem_times_ms.csv", index=False)

# --- Visualizations ---
normal_idx = np.where(labels == 0)[0]
noisy_idx = np.where(labels == 1)[0]

plt.figure(figsize=(8,5))
for idx in np.random.default_rng(0).choice(normal_idx, size=min(100, len(normal_idx)), replace=False):
    plt.plot(t_ms, curves[idx], alpha=0.4)
plt.xscale("log")
plt.xlabel("Time (ms)"); plt.ylabel("Amplitude (arb. units)")
plt.title("Sample of NORMAL synthetic TDEM decays (overlay)")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,5))
for idx in noisy_idx:
    plt.plot(t_ms, curves[idx], alpha=0.7)
plt.xscale("log")
plt.xlabel("Time (ms)"); plt.ylabel("Amplitude (arb. units)")
plt.title("NOISY synthetic TDEM decays (overlay)")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,5))
ex_normal = np.random.default_rng(1).choice(normal_idx)
ex_noisy = np.random.default_rng(2).choice(noisy_idx)
plt.plot(t_ms, curves[ex_normal], label="normal", alpha=0.9)
plt.plot(t_ms, curves[ex_noisy], label="noisy", alpha=0.9)
plt.xscale("log")
plt.xlabel("Time (ms)"); plt.ylabel("Amplitude (arb. units)")
plt.title("Example: one normal vs one noisy curve")
plt.legend()
plt.tight_layout()
plt.show()

print("Saved: /mnt/data/tdem_synth.csv, /mnt/data/tdem_synth.parquet (if available), /mnt/data/tdem_times_ms.csv")
'''

# Create the notebook
try:
    import nbformat as nbf
    nb = nbf.v4.new_notebook()
    nb['cells'] = [nbf.v4.new_code_cell(nb_code)]
    notebook_path = "/mnt/data/tdem_synthetic.ipynb"
    with open(notebook_path, "w", encoding="utf-8") as f:
        nbf.write(nb, f)
    created_nb = notebook_path
except Exception as e:
    created_nb = None
    print("Warning: could not create .ipynb file:", e)

# Final paths to report back
print("CSV path:", csv_path)
print("Parquet path:", parquet_path if parquet_path else "Parquet not created")
print("Times CSV path:", times_csv_path)
print("Notebook path:", created_nb if created_nb else "Notebook not created")
```
